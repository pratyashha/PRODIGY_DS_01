```python
import pandas as pd

# Load the dataset
data = pd.read_csv(r"C:\Users\PRATYASHA TRIPATHY\Internship\Prodigy\PRODIGY_DS_01\data2.csv")

# Display the first few rows of the dataset
print(data.head())
```

```python
# Check for missing values
print(data.isnull().sum())

# Drop rows with missing values in key columns (optional)
data_clean = data.dropna(subset=['Country Code', 'Region', 'IncomeGroup'])

# Display the cleaned data
print(data_clean.head())
```

```python
import matplotlib.pyplot as plt

# Count the number of occurrences for each region
region_counts = data_clean['Region'].value_counts()

# Create a bar chart
plt.figure(figsize=(10, 6))
region_counts.plot(kind='bar', color='skyblue', edgecolor='k')
plt.xlabel('Region')
plt.ylabel('Count')
plt.title('Distribution of Regions')
plt.xticks(rotation=45)
plt.show()
```

```python
# Count the number of occurrences for each income group
income_counts = data_clean['IncomeGroup'].value_counts()

# Create a bar chart
plt.figure(figsize=(10, 6))
income_counts.plot(kind='bar', color='lightgreen', edgecolor='k')
plt.xlabel('Income Group')
plt.ylabel('Count')
plt.title('Distribution of Income Groups')
plt.xticks(rotation=45)
plt.show()
```

```python
# Check if the population column exists and create a histogram
if 'Population, total' in data.columns:
    plt.figure(figsize=(10, 6))
    plt.hist(data['Population, total'].dropna(), bins=30, edgecolor='k')
    plt.xlabel('Population')
    plt.ylabel('Frequency')
    plt.title('Distribution of Population')
    plt.show()
else:
    print("Population data is not available in the dataset.")
```
